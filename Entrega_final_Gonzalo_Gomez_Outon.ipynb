##üß© Etapa 2: Preprocesamiento y Limpieza de Datos
**Objetivo:** Demostrar conocimiento de las t√©cnicas de limpieza y transformaci√≥n de datos.

### 1Ô∏è‚É£ Limpieza de Datos: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados. Documentar el proceso y los resultados.

```python
# TODO: Limpiar el conjunto de datos eliminando duplicados y caracteres no deseados.
# Sugerencia: aplicar .drop_duplicates(), .str.strip(), .str.replace() seg√∫n el caso.
# Documentar el proceso y los resultados.
# Eliminar nulos

# ============================================
# LIMPIEZA Y NORMALIZACI√ìN DE LOS DATASETS
# ============================================
# Se limpian y normalizan los DataFrames:
#   ventas, clientes, marketing
# ============================================

# -------------------------------------------------
# 1Ô∏è‚É£ Crear copias independientes para no modificar los originales
# -------------------------------------------------
ventas_clean   = ventas.copy()
clientes_clean = clientes.copy()
marketing_clean = marketing.copy()

print("Copias creadas correctamente.")
```

```python
# -------------------------------------------------
# 2Ô∏è‚É£ Eliminar filas completamente duplicadas
# -------------------------------------------------
ventas_clean   = ventas_clean.drop_duplicates()
clientes_clean = clientes_clean.drop_duplicates()
marketing_clean = marketing_clean.drop_duplicates()

print("Duplicados eliminados.")
print("Ventas:", ventas_clean.shape, "Clientes:", clientes_clean.shape, "Marketing:", marketing_clean.shape)
```

```python
#Revisi√≥n post limpieza
print("Control de calidad luego de limpiar duplicados:\n")

print("VENTAS - Duplicados:", ventas_clean.duplicated().sum())
print("CLIENTES - Duplicados:", clientes_clean.duplicated().sum())
print("MARKETING - Duplicados:", marketing_clean.duplicated().sum())
```

```python
# -------------------------------------------------
# 3Ô∏è‚É£ Funci√≥n para limpiar texto en columnas tipo string
# -------------------------------------------------
def normalizar_texto(df):
    for col in df.select_dtypes(include="object").columns:
        df[col] = (
            df[col]
            .astype(str) #convierte a string
            .str.strip() #quita espacios al inicio/fin
            .str.replace(r"[\u200b\t\r\n]", "", regex=True) #borra caracteres invisibles
            .str.replace(" +", " ", regex=True) #reemplaza espacios m√∫ltiples por uno
            .str.title() #pone may√∫scula inicial a cada palabra
        )
    return df

# Aplicamos la funci√≥n a los tres DataFrames
ventas_clean = normalizar_texto(ventas_clean)
clientes_clean = normalizar_texto(clientes_clean)
marketing_clean = normalizar_texto(marketing_clean)

print("Textos normalizados.")
```

```python
# -------------------------------------------------
# 4Ô∏è‚É£Normalizar fechas
# -------------------------------------------------
# Si alguna columna contiene fechas (por ejemplo "fecha" o "fechanotificacion"),
# se intenta convertir a formato datetime de pandas.
# to_datetime intenta interpretar el formato y transforma valores inv√°lidos en NaT (Not a Time).

for df in [ventas_clean, clientes_clean, marketing_clean]:
    for col in df.columns:
        if "fecha" in col.lower():  # detecta columnas con la palabra "fecha"
            df[col] = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
            # Par√°metros:
            #   errors="coerce" ‚Üí convierte valores no v√°lidos en NaT (evita error)
            #   dayfirst=True   ‚Üí interpreta formatos tipo "DD/MM/YYYY" (formato latino)
#n

print("Fechas convertidas correctamente.")

print(ventas_clean.dtypes)
print(clientes_clean.dtypes)
print(marketing_clean.dtypes)
```

```python
# -------------------------------------------------
# 5Ô∏è‚É£ Aplicar la normalizaci√≥n de texto
# -------------------------------------------------
ventas_clean = normalizar_texto(ventas_clean)
clientes_clean = normalizar_texto(clientes_clean)
marketing_clean = normalizar_texto(marketing_clean)

#mostramos los df luego de normalizar los textos para revisar que queden bien
print(ventas_clean.head(5))
print(clientes_clean.head(5))
print(marketing_clean.head(5))
```

```python
# 6Ô∏è‚É£ Normalizar valores num√©ricos (por ejemplo: precio)
# -------------------------------------------------
if "precio" in ventas_clean.columns:
    ventas_clean["precio"] = (
        ventas_clean["precio"]
        .astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    ventas_clean["precio"] = pd.to_numeric(ventas_clean["precio"], errors="coerce")

print("Columnas num√©ricas normalizadas (si exist√≠an s√≠mbolos o texto).")
```

### 2Ô∏è‚É£ Transformaci√≥n de Datos: Aplicar filtros y transformaciones para crear una tabla de ventas que muestre solo los productos con alto rendimiento.

```python
# 2) Transformaci√≥n de Datos ‚Äî Productos de alto rendimiento

ventas_perf = ventas_clean.assign(
    ingreso = ventas_clean["precio"] * ventas_clean["cantidad"]
)

print("\n=== Vista previa: ventas_perf con columna 'ingreso' ===")
display(ventas_perf.head())

resumen_prod = (
    ventas_perf
    .groupby("producto", as_index=False)
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        precio_promedio=('precio', 'mean'),
        registros=('ingreso', 'size')
    )
)

print("\n=== Resumen por producto ===")
display(resumen_prod.head())

p80_ingreso = resumen_prod["ingreso_total"].quantile(0.80)

ventas_top = (
    resumen_prod
    .query("ingreso_total >= @p80_ingreso")
    .sort_values(["ingreso_total", "unidades"], ascending=False)
)

print("\n=== Productos de alto rendimiento (TOP 20%) ===")
display(ventas_top)

```

### 3Ô∏è‚É£ Agregaci√≥n: Resumir las ventas por categor√≠a de producto y analizar los ingresos generados.

```python
# TODO: Resumir las ventas por categor√≠a de producto y analizar los ingresos generados.
# Sugerencia: usar .groupby() y .agg() para generar m√©tricas como suma y promedio.
ventas_cat = ventas_perf.copy()
# Aseguramos que 'cantidad' sea entero
ventas_cat["cantidad"] = (
    pd.to_numeric(ventas_cat["cantidad"], errors="coerce")
    .fillna(0)
    .astype(int)
)

resumen_cat = (
    ventas_cat
    .groupby("categoria", as_index=False)
    .agg(
        ingreso_total=('ingreso', 'sum'),
        unidades=('cantidad', 'sum'),
        ventas=('ingreso', 'size'),
        precio_promedio=('precio','mean')
    )
    .assign(ticket_promedio_por_venta=lambda df: df["ingreso_total"] / df["ventas"])
    .sort_values("ingreso_total", ascending=False, ignore_index=True)
)

print("\n=== Resumen de ventas por categor√≠a ===")
display(resumen_cat)
```

Las tres categor√≠as muestran niveles de ventas bastante parecidos en cantidad de unidades y transacciones.
La diferencia principal est√° en el valor promedio por venta, lo cual afecta directamente al ingreso total.

Electrodom√©sticos lidera, tanto en ventas como en valor por ticket.

Electr√≥nica sigue de cerca, con m√©tricas muy similares.

Decoraci√≥n vende casi lo mismo en volumen, pero genera menos ingreso porque sus productos tienen precios m√°s bajos.

Recomendaci√≥n:
Si el objetivo del negocio es crecer en ingreso, conviene potenciar Decoraci√≥n (que tiene buena demanda pero bajo ticket), o invertir m√°s en Electrodom√©sticos donde el retorno por venta es mayor.

### 4Ô∏è‚É£ Integraci√≥n de Datos: Combinar los sets de datos de ventas y marketing para obtener una visi√≥n m√°s amplia de las tendencias.

```python
# TODO: Combinar los sets de datos de ventas y marketing para obtener una visi√≥n m√°s amplia de las tendencias.
# Sugerencia: usar pd.merge() especificando la clave com√∫n entre ambos DataFrames.
# Documentar cualquier observaci√≥n relevante sobre la combinaci√≥n de datos.
# 4) Integraci√≥n de Datos ‚Äî Ventas + Marketing

# Ingreso total por producto
df_ventas_x_prod = (
    ventas_clean
    .assign(ingreso=lambda x: x["precio"] * x["cantidad"])
    .groupby("producto", as_index=False)
    .agg(precio_por_cantidad=("ingreso", "sum"))
)

print("\n=== Ingreso total por producto ===")
display(df_ventas_x_prod.head())

#Costo total por producto
df_marketing_x_prod = (
    marketing_clean
    .groupby("producto", as_index=False)
    .agg(costo=("costo", "sum"))
)

print("\n=== Costo de marketing por producto ===")
display(df_marketing_x_prod.head())

#Uni√≥n
merged = pd.merge(df_ventas_x_prod, df_marketing_x_prod, on="producto", how="left")

merged["costo_sobre_ingreso_%"] = (
    merged["costo"] / merged["precio_por_cantidad"] * 100
)

print("\n=== Integraci√≥n Ventas + Marketing ===")
display(merged.head())
```

##üß© Etapa 3: An√°lisis de Datos

### 1Ô∏è‚É£ Estad√≠stica Descriptiva: Calcular medidas de tendencia central y dispersi√≥n sobre las ventas. Presentar los resultados en un informe.

```python
# 1) Estad√≠stica descriptiva de ingresos por producto

df_ing = resumen_prod.copy()

media    = df_ing['ingreso_total'].mean()
mediana  = df_ing['ingreso_total'].median()
moda     = df_ing['ingreso_total'].mode()[0]

rango      = df_ing['ingreso_total'].max() - df_ing['ingreso_total'].min()
varianza   = df_ing['ingreso_total'].var()
desviacion = df_ing['ingreso_total'].std()

q1 = df_ing['ingreso_total'].quantile(0.25)
q3 = df_ing['ingreso_total'].quantile(0.75)
iqr = q3 - q1

lim_inf = q1 - 1.5 * iqr
lim_sup = q3 + 1.5 * iqr

outliers = df_ing[(df_ing['ingreso_total'] < lim_inf) | (df_ing['ingreso_total'] > lim_sup)]

print("\n=== ESTAD√çSTICAS DESCRIPTIVAS ===")
print(f"Media: {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Moda: {moda:.2f}")
print(f"Rango: {rango:.2f}")
print(f"Varianza: {varianza:.2f}")
print(f"Desviaci√≥n est√°ndar: {desviacion:.2f}")
print(f"IQR: {iqr:.2f}")
print(f"L√≠mite inferior para outliers: {lim_inf:.2f}")
print(f"L√≠mite superior para outliers: {lim_sup:.2f}")

print("\n=== OUTLIERS DETECTADOS ===")
display(outliers)

print("\n=== Descripci√≥n estad√≠stica (describe) ===")
display(df_ing['ingreso_total'].describe())
```

### 3Ô∏è‚É£ Correlaci√≥n: Evaluar la relaci√≥n entre diferentes variables (por ejemplo, precio y cantidad vendida) y mostrar los resultados en un gr√°fico.

```python
x = resumen_prod["precio_promedio"]
y = resumen_prod["unidades"]

# Ajuste lineal
m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.7)
plt.plot(x, m*x + b, color="red", linewidth=2, label="Tendencia")

plt.title("Relaci√≥n Precio Promedio y Unidades Vendidas (por producto)", fontsize=14)
plt.xlabel("Precio Promedio ($)", fontsize=12)
plt.ylabel("Unidades Vendidas", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Coeficiente de correlaci√≥n (precio_promedio vs unidades)
# Calculamos el coeficiente de correlaci√≥n de Pearson con pandas (m√©todo corr)
corr_val = resumen_prod[['precio_promedio','unidades']].corr().loc['precio_promedio','unidades']

print(f"Coeficiente de correlaci√≥n (Pearson): {corr_val:.3f}")
# Interpretaci√≥n simple:
if corr_val > 0.5:
    print("Interpretaci√≥n: relaci√≥n positiva clara (a mayor precio_promedio, suelen venderse m√°s unidades).")
elif corr_val > 0.2:
    print("Interpretaci√≥n: relaci√≥n positiva d√©bil a moderada.")
elif corr_val < -0.5:
    print("Interpretaci√≥n: relaci√≥n negativa clara (a mayor precio_promedio, suelen venderse menos unidades).")
elif corr_val < -0.2:
    print("Interpretaci√≥n: relaci√≥n negativa d√©bil a moderada.")
else:
    print("Interpretaci√≥n: relaci√≥n muy d√©bil o nula.")
```

### 1Ô∏è‚É£ Visualizaci√≥n con Matplotlib: Crear gr√°ficos b√°sicos (l√≠neas, barras, dispersi√≥n)

```python
# =========================================
# Tendencia real de ingresos a lo largo del tiempo
# =========================================
# Usamos el dataset 'ventas_perf', que contiene la fecha y el ingreso de cada registro.

import matplotlib.pyplot as plt
import pandas as pd

# Convertimos la columna 'fecha_venta' a tipo datetime (por seguridad)
ventas_perf['fecha_venta'] = pd.to_datetime(ventas_perf['fecha_venta'], errors='coerce')

# Agrupamos las ventas por fecha y sumamos los ingresos de cada d√≠a
ingresos_por_dia = ventas_perf.groupby('fecha_venta')['ingreso'].sum().sort_index()

# Calculamos el acumulado de ingresos
ingresos_acumulados = ingresos_por_dia.cumsum()

# Graficamos ambas curvas: ingresos diarios y acumulados
plt.figure(figsize=(10, 5))

# L√≠nea 1: ingresos diarios
plt.plot(ingresos_por_dia, color='skyblue', label='Ingreso diario')

# L√≠nea 2: ingresos acumulados
plt.plot(ingresos_acumulados, color='royalblue', label='Ingreso acumulado', linewidth=2)

# Personalizaci√≥n del gr√°fico
plt.title('Evoluci√≥n de ingresos por fecha de venta')
plt.xlabel('Fecha de venta')
plt.ylabel('Ingresos ($)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# =========================================
# Barras: Ingresos Totales por Producto
# =========================================

import matplotlib.pyplot as plt

df = resumen_prod.copy()
df_ordenado = df.sort_values('ingreso_total', ascending=False)

plt.figure(figsize=(14, 7))

color_principal = "#3A78D6"

barras = plt.bar(
    df_ordenado['producto'],
    df_ordenado['ingreso_total'],
    color=color_principal,
    edgecolor='black',
    linewidth=0.6
)

plt.title("Ingresos Totales por Producto", fontsize=16, weight="bold")
plt.xlabel("Producto", fontsize=12)
plt.ylabel("Ingreso Total ($)", fontsize=12)
plt.xticks(rotation=90)

# Etiquetas dentro de cada barra
for b in barras:
    h = b.get_height()
    x_c = b.get_x() + b.get_width() / 2
    plt.text(
        x_c,
        h * 0.5,
        f"${h:,.0f}",
        ha='center',
        va='center',
        color="white",
        fontsize=9,
        rotation=90
    )

plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
```
